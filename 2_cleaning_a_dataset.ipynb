## Cleaning up the price

In this sectino we are going to transform the `price` column into a float column in USD dolars.

For this, be careful of the different formats in the data.

For simplicity, you might assume that all the prices are in USD dolars, regarding of the symbol of the currency used.

```python
def clean_price(price):
    """
    In this function we will transform the
    `price` column into a column of floats.
    In case a product has more than one price,
    return the lowest one.
    """
    # Remove currency symbols, commas, and "USD"
    price = price.replace('$', '').replace('€', '').replace('£', '').replace('¥', '').replace('₹', '').replace('USD', '')
    price = price.replace(',', '')  # Remove thousands separators

    # Replace em dash or en dash with a standard hyphen
    price = price.replace('–', '-')  # This line is added to replace the em dash
    price = price.replace('—', '-')  # This line is added to replace the en dash

    # Remove newline and tab characters, and non-breaking spaces
    price = price.replace('\n', '').replace('\t', '').replace('\xa0', ' ')

    # Extract the first numerical value from the string
    import re
    match = re.search(r'(\d+\.?\d*)', price) # Find the first numerical value
    if match:
        price = match.group(1)
    else:
        return float('nan') # Handle cases with no numerical value

    # Check if there's a range and split to get the minimum value
    if '-' in price:
        min_price = float(price.split('-')[0].strip()) # Changed this line to handle price range
    else:
        min_price = float(price.strip())
    return min_price

# Apply the function to each row in the 'price' column
df['price'] = df['price'].apply(clean_price)
```

```python
def low_high_product_mean(df):
    """
    Finally, we will calculate `product_category` with the lowest and highest mean price
    for the brand Victoria's Secret.
    """
    df2=df[df["brand_name"]=="victoria's secret"]

    aggregated_data = df2.groupby('product_category')['price'].agg(['min', 'max', 'mean'])

    lowest_price_category=aggregated_data['mean'].min()
    highest_price_category=aggregated_data['mean'].max()

    return lowest_price_category, highest_price_category


    # Write your code here
```
